**[Abalone Data Set](https://archive.ics.uci.edu/ml/datasets/Abalone)**

```python
import pandas as pd 
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names',
                 header=None,
                 names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country'])
```

# Step 6
### Plot the distribution of three types of Abalone using different colors where x-axis is the diameter and y-axis the whole weight

```python
%matplotlib inline
import matplotlib.pyplot as plt

colorDict = {
    'M': 0,
    'F': 1,
    'I': 2
}

plt.xlabel('Diameter')
plt.ylabel('whole weight')
plt.scatter(df['Diameter'], df['whole weight'], c = [colorDict[c] for c in df['Sex']])
```

# Step 7
### Show the proportion of three types of Abalone using pie chart.

```python
plt.pie(
    [len(df[df['Sex'] == 'M']), len(df[df['Sex'] == 'F']), len(df[df['Sex'] == 'I'])],
    explode=[0.05, 0, 0],
    labels = ['M', 'F', 'I'],
    autopct= "%3.1f%%")
```
